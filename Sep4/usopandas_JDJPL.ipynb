# El uso de algunas funciones de pandas

Se toman los datos de Temperatura promedio en estados de México en los años 2018 y 2019, de la fuente: https://smn.conagua.gob.mx/es/climatologia/temperaturas-y-lluvias/resumenes-mensuales-de-temperaturas-y-lluvias
Aplicando funciones se filtran datos de interés.

```python
import os
import requests
import pandas as pd
```

```python
TM2018=pd.read_csv('TMed2018.csv')
TM2018D=pd.DataFrame(TM2018)
#TM2018D
TABLA2018=pd.DataFrame({'EDO':TM2018D['Edo'],'TM2018':TM2018D['Tmed']})
TABLA2018
```

```python
PROMEDIO2018=TABLA2018.mean()
PROMEDIO2018
```

```python
TM2019=pd.read_csv('TMed2019.csv')
TM2019D=pd.DataFrame(TM2019)
#TM2019
TABLA2019=pd.DataFrame({'EDO':TM2019D['Edo'],'TM2019':TM2019D['Tmed']})
TABLA2019
```

```python
PROMEDIO2019=TABLA2019.mean()
PROMEDIO2019
```

En el siguiente punto se unen los datos de ambas tablas (datos 2018 y 2019).

```python
pd.concat([TM2018D, TM2019D], ignore_index = True ) #Se unen los datos de ambos años, ignore_index= True *Para enumerar nuevamente
```

En la siguiente entrada se unen los datos de 2018 y 2019 referentes a las temperaturas, obviamente esto no es muy útil, puesto que resulta en una conbinación elemento-elementos. 

```python
DATOS=pd.merge(TABLA2018,TABLA2019)
DATOS 
```

Suponiendo que nos interesa conocer el promedio de las temperaturas promedio en el año 2019 de cada uno de los estados reportados.

```python
TABLA2019.groupby('EDO')['TM2019'].median()
```

Para saber la cantidad de datos de cada estado, se hace:

```python
TABLA2019.groupby('EDO')['TM2019'].count()
```

```python
TABLA2019.groupby(['EDO', 'TM2019'])
```

Ahora se agrupan formalmente los datos (2019) de acuerdo al estado:

```python
group_a = TABLA2019.groupby('EDO')  #Agraupar en estados
for estados, temperaturas in group_a: #Iterar sobre la agrupación establecida para desplegar grupos
    print(estados)
    print(temperaturas)
```

Si se necesita filtrar aún más la información, en este caso extraer los datos únicamente del estado Guanajuato, hacer:

```python
data_GTO = group_a.get_group('GTO')
print(data_GTO)
```
