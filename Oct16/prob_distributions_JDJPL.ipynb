## Se estudian varias funciones de densidad de probabilidad (PDF) así como sus funciones de distribución acumulada de probabilidad (CDF) de gran importancia. 

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import math
import scipy.special as sc
import scipy.integrate as integrate
```

# Chi-Squared distribution

```python
k=2
import random
n=200 #Tamaño de la muestra
x=np.zeros(n)
pdf=np.zeros(n)
cdf=np.zeros(n)
for i in range(n):
    #x[i]=random.random()  ##.randint(,)
    x[i]=random.uniform(0, 10) #generadon valores aleatorios entre el intervalo de -1 a 1.
    pdf[i]=(.5**(k/2)/math.gamma(x[i]))*x[i]**(.5*k-1)*np.exp(-x[i]/2)
    cdf[i]=sc.gammainc(k/2,x[i]/2)/math.gamma(k/2)
```

```python
plt.plot(x,pdf,'.')
plt.title('Función de Densidad de Probabilidad ji-cuadrado')
plt.show()
```

```python
plt.plot(x,cdf,'.')
plt.title('Función de Densidad de Probabilidad ji-cuadrado')
plt.show()
```

# Poisson distribution
k: Número de ocurrencias del evento o fenómeno (la función nos da la probabilidad de que el evento suceda precisamente k veces).

L: Es un parámetro positivo que representa el número de veces que se espera que ocurra el fenómeno durante un intervalo dado.

```python
k=1
import random
n=200 #Tamaño de la muestra
L=np.zeros(n)
pf=np.zeros(n)
cdf=np.zeros(n)
for i in range(n):
    #x[i]=random.random()
    L[i]=random.uniform(0, 10) #generadon valores aleatorios entre el intervalo de -1 a 1.
    pf[i]=np.exp(-L[i])*L[i]**k/math.factorial(k)
    cdf[i]=sc.gammainc(k+1,L[i])/math.factorial(k)
```

```python
plt.plot(L,pf,'.')
plt.title('o')
plt.show()
```

```python
plt.plot(L,cdf,'.')
plt.title('Función de Densidad de Probabilidad ji-cuadrado')
plt.show()
```

# Beta distribution
k: Número de ocurrencias del evento o fenómeno (la función nos da la probabilidad de que el evento suceda precisamente k veces).

L: Es un parámetro positivo que representa el número de veces que se espera que ocurra el fenómeno durante un intervalo dado.

```python
a=2
b=5
import random
n=200 #Tamaño de la muestra
x=np.zeros(n)
pdf=np.zeros(n)
cdf=np.zeros(n)
for i in range(n):
    #x[i]=random.random()
    x[i]=random.uniform(0, 1) #generadon valores aleatorios entre el intervalo de -1 a 1.
    pdf[i]=math.gamma(a+b)*x[i]**(a-1)*(1-x[i])**(b-1)
    I=integrate.quad(lambda t: math.gamma(a+b)*t**(a-1)*(1-t)**(b-1) , 0, x[i])
    cdf[i]=I[0]-I[1]
```

```python
plt.plot(x,pdf,'.')
plt.title('Función de densidad de probabilidad para la distribución Beta')
plt.show()
```

```python
plt.plot(x,cdf,'.')
plt.title('Función de distribución de probabilidad acumulada para la distribución Beta')
plt.show()
```

# Binomial distribution

```python
p=.5
import random
n=30 #Tamaño de la muestra
x=np.zeros(n)
pf=np.zeros(n)
cdf=np.zeros(n)
for i in range(n):
    #x[i]=random.random()
    x[i]=random.randint(0, n) #generadon valores aleatorios entre el intervalo de 0 a 10.
    pf[i]=(math.factorial(n)/(math.factorial(x[i])*math.factorial(n-x[i])))*(p**x[i])*(1-p)**(n-x[i])

   
```

```python
plt.plot(x,pf,'o')
plt.title('Función de probabilidad para la distribución Binomial')
plt.show()
```

```python
for i in range(n):
    S=0
    for j in range(n):
        if x[j] <= x[i]:
            S=S+(math.factorial(n)/(math.factorial(x[j])*math.factorial(n-x[j])))*(p**x[j])*(1-p)**(n-x[j])
    cdf[i]=S
```

```python
plt.plot(x,cdf,'o')
plt.title('CDF para la distribución Binomial')
plt.show()
```
