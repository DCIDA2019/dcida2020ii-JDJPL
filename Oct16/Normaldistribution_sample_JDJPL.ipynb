Generate as many as you want, and make the histogram of H0 to verify you did it right. Use a mean of 70 and a sigma=2.
Para mostrar los resultados de manera gráfica se tomarán valores aleatorios en el rango de [-1,1], de un tamaño de muestra de n elementos. Además se considera una media de 0 y una desviación típica de 0.2.

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import math
```

```python
mu=0 #Media
s=0.2**(.5) #Desviación típica **.5
import random
n=500 #Tamaño de la muestra
x=np.zeros(n)
f=np.zeros(n)
cdf=np.zeros(n)
for i in range(n):
    #x[i]=random.random()
    x[i]=random.uniform(-1, 1) #generadon valores aleatorios entre el intervalo de -1 a 1.
    f[i]=((s*(2*np.pi)**(1/2))**(-1))*np.exp( -(x[i]-mu)**(2)/(2*s**2) )
    cdf[i]=.5*(1+math.erf((x[i]-mu)/(s*2**(.5))))
```

```python
plt.plot(x,f,'.')
plt.title('Función de Densidad de Probabilidad')
plt.show()
```

```python
plt.plot(x,cdf,'.')
plt.title('Función de distribución acumulada')
plt.show()
```
