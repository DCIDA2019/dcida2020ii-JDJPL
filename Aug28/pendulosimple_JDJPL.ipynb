## Solución al péndulo simple

En el siguiente Jupyter Notebook se desarrolla un código didáctico para resolver la ecuación de movimiento del péndulo simple y que servirá como actividad con fecha 28-agosto-2020 para la materia Análisis de Datos. 

La ecuación de movimiento que rige al péndulo simple es: $\ddot\theta+\frac gl\sin\theta=0$. Donde $g$ representa es el valor de la gravedad y $l$ la longitud del péndulo. Dado que es una ecuación diferencial de segundo orden se tiene que especificar las condiciones iniciales: $\theta(0)=\theta_0$ y $\dot\theta(0)={\dot\theta}_0$.

1. Se escribe el código usando sintaxis básica.

```python
import numpy as np
import matplotlib.pyplot as plt
import timeit
```

```python
#Constantes y parámetros del sistema
l=.2 #Longitud del péndulo
k=-9.81/l 
t0=0 #Tiempo inicial
tf=10 #Tiempo final
n=500 #Partición
```

```python
start = timeit.timeit()
dt=(tf-t0)/n #Salto temporal
t=[None]*(n)
x=[None]*(n)
v=[None]*(n)
#Vector valores de t
for jj in range(0,n-1):
    t[jj]=t0+dt*jj
#Vectores para almacenar valores
for j in range(0,n-1): 
    x[j]=0
    v[j]=0
#Condiciones iniciales
x[0]=1.57
v[0]=2
#Implementación de Runge-Kutta
#Iteraciones en las soluciones del sistema en función del tiempo
for i in range(0,n-1):
    k1=dt*v[i]
    l1=dt*(np.sin(x[i]))*k
    k2=dt*(v[i]+l1/2)
    l2=dt*(np.sin(x[i]+k1/2))*k
    k3=dt*(v[i]+l2/2)
    l3=dt*(np.sin(x[i]+k2/2))*k
    k4=dt*(v[i]+l3)
    l4=dt*(np.sin(x[i]+k3))*k
    
    x[i+1]=x[i]+(k1+2*k2+2*k3+k4)/6
    v[i+1]=v[i]+(l1+2*l2+2*l3+l4)/6
    
end = timeit.timeit() 
print(end-start) #Calcula el tiempo de ejecución
```

```python
#Resultados. Gráfica de la posición angular de la masa respecto al tiempo.
fig=plt.figure(1)
ax1=fig.add_subplot(111)
ax1.plot(t,x,color="blue",linewidth=1,linestyle="-",fillstyle="none")    
ax1.set_xlabel("$t$ [s] ",color="blue")        
ax1.set_ylabel("$x$ [rad]",color='b')
plt.show()
```

```python
#Resultados. Gráfica de la velocidad angular de la masa respecto al tiempo.
fig=plt.figure(1)
ax1=fig.add_subplot(111)
ax1.plot(t,v,color="b",linewidth=1,linestyle="-",fillstyle="none")    
ax1.set_xlabel("$t$ [s] ",color="b")        
ax1.set_ylabel("$v$ [rad/s]",color='b')
plt.show()
```

2. Ahora se reescribe el código utilizando arreglos de la librería numpy.

```python
start = timeit.timeit()
dt=(tf-t0)/n #Salto temporal
#Vector valores de t
t=np.arange(t0,tf,dt)
#Vectores para almacenar valores
x=np.zeros(n) 
v=np.zeros(n) 
#Condiciones iniciales
x[0]=1.57
v[0]=2
#Implementación de Runge-Kutta
#Iteraciones en las soluciones del sistema en función del tiempo
for i in range(0,n-1):
    k1=dt*v[i]
    l1=dt*(np.sin(x[i]))*k
    k2=dt*(v[i]+l1/2)
    l2=dt*(np.sin(x[i]+k1/2))*k
    k3=dt*(v[i]+l2/2)
    l3=dt*(np.sin(x[i]+k2/2))*k
    k4=dt*(v[i]+l3)
    l4=dt*(np.sin(x[i]+k3))*k
    
    x[i+1]=x[i]+(k1+2*k2+2*k3+k4)/6
    v[i+1]=v[i]+(l1+2*l2+2*l3+l4)/6
    
end = timeit.timeit()
print(end-start) #Calcula el tiempo de ejecución
```

```python
#Resultados. Gráfica de la posición angular de la masa respecto al tiempo.
fig=plt.figure(1)
ax1=fig.add_subplot(111)
ax1.plot(t,x,color="g",linewidth=1,linestyle="-",fillstyle="none")    
ax1.set_xlabel("$t$ [s] ",color="g")        
ax1.set_ylabel("$x$ [rad]",color='g')
plt.show()
```

```python
#Resultados. Gráfica de la velocidad angular de la masa respecto al tiempo.
fig=plt.figure(1)
ax1=fig.add_subplot(111)
ax1.plot(t,v,color="g",linewidth=1,linestyle="-",fillstyle="none")    
ax1.set_xlabel("$t$ [s] ",color="g")        
ax1.set_ylabel("$v$ [rad/s]",color='g')
plt.show()
```

Nota: Se calcula el tiempo de ejecución cuando se generan los vectores asociados al tiempo, posición y velocidad, y se resuelve la ecuación diferencial numéricamente.
